## Reading Data Values

```python
import json
fd = open("Inventory.json",'r')
r = fd.read()
fd.close()

inv = json.loads(r)
```

```python
inv
```

```python
prod_id = int(input("Enter Product id:"))
name=str(input("Enter Name:"))
price=int(input("Enter Price:"))
quan=int(input("Enter Quantity:"))
category=str(input("Enter Product Category:"))
pp=int(input("Enter Product Popularity:"))

inv[prod_id] = {'Name': name, 'Price': price, 'Quantity': quan, 'Category': category, 'Product Popularity': pp}

#Writing the data into inventory
js = json.dumps(inv)
fd = open("Inventory.json",'w')
fd.write(js)
fd.close()
```
